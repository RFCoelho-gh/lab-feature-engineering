There are multiple strategies to handle missing data

- Removing all rows or all columns containing missing data.
- Filling all missing values with a value (mean in continouos or mode in categorical for example).
- Filling all missing values with an algorithm.

For this exercise, because we have such low amount of null values, we will drop rows containing any missing value.

- Drop PassengerId and Name

```python
#your code here

spaceship = spaceship.drop(columns=['PassengerId', 'Name'])
```

- For non-numerical columns, do dummies.

```python
#your code here

# Convert non-numerical columns to dummy variables
spaceship_dummies = pd.get_dummies(spaceship, columns=['HomePlanet', 'CryoSleep','Cabin','Destination'])
```

```python
spaceship_dummies
```

**Perform Train Test Split**

```python
spaceship_dummies.columns
```

```python
features = spaceship_dummies.drop(columns=['Transported'])
```

```python
target = spaceship_dummies["Transported"]
```

```python
#your code here

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.20, random_state=0)
```

```python
## The code below takes all numerical (>3) columns and normalizes them, in order to avoid range distortion

# Identify columns with a maximum value above 3
columns_to_normalize = [col for col in spaceship_dummies.columns if spaceship_dummies[col].max() > 10]
```

```python
# Initialize scaler and fit only on training data
scaler = MinMaxScaler()
X_train[columns_to_normalize] = scaler.fit_transform(X_train[columns_to_normalize])
```

```python
# Apply the scaler transformation to the test data
X_test[columns_to_normalize] = scaler.transform(X_test[columns_to_normalize])
```
